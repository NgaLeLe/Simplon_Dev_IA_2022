```python
survival = pd.read_csv("donnees/titanic_survival.csv")
c_age = survival['age']
age_null = c_age.isnull()
#nb_true = age_null[age_null==True].count()
nb_false = age_null[age_null==False].count()
nb_true = len(age_null[c_age.isnull()])
print("Null = ", nb_true)
print("Not null = ", nb_false)
```

```python
#probleme de valeur null
#fait un mask isnull s/colonne age
age_isnull = pd.isnull(survival["age"])
#creer nv serie qui contient 
good_age = survival["age"][age_isnull == False]
mean_age = sum(good_age) /len(good_age)
print("age de moyen = ", mean_age)
```

```python
fare_null = pd.isnull(survival["fare"])
bon_fare = survival["fare"][fare_null == False]
mean_fare = sum(bon_fare) /len(bon_fare)
print("Fare moyenne = ", mean_fare)
```
