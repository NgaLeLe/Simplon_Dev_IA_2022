```python
#applique des fonc s/dataframe
def null_count(column) :
    #faire un mask isnull s/column
    colnull = pd.isnull(column)
    null = column[colnull]
    #apliquer mask s/col, suite stocker dans var null 
    return len(null)

#apply fonction null_count s tous les colonne de DF
col_null_cnt = survival.apply(null_count)
print(col_null_cnt)
```

```python
#applique methods s/list de colonne
#choisir des colonnes [list_col]
col_age_null_cnt = survival[["pclass", "age"]].apply(null_count)
print(col_age_null_cnt)
```

```python
#applique une fonction à 1 ligne
def gen_type_age(column) :
    tmp_ans = ""
    #mnull = pd.isnull(column) 
    if pd.isnull(column["age"])== True :
        tmp_ans = "unknown"
    elif  column["age"] < 18:
        tmp_ans = "mineur"
    else :
        tmp_ans = "adult"
    return tmp_ans
#axis = 1 -> ç'applique sur ligne par ligne   
type_age = survival.apply(gen_type_age, axis = 1)
print(type_age)
```

```python
def type_age(ma_ligne):
    if pd.isnull(ma_ligne["age"]) == True :
        return "unknown"
    elif ma_ligne["age"] < 18 :
        return "minor"
    else :
        return "adult"

survival.apply(type_age, axis=1)
```

```python
#operation s/les colonnes
sodium_gr  = iffood["Sodium_(mg)"] /1000
surce_mgr = iffood["Sugar_Tot_(g)"] * 1000
print("sodium_gr = ", sodium_gr)
print("surce_mgr = ", surce_mgr)
```

```python
gr_protein_pr_water = iffood["Protein_(g)"] /iffood["Water_(g)"]
mlgr_cal_iron = iffood["Calcium_(mg)"] + iffood["Iron_(mg)"]
print("gr_protein_pr_water = ", gr_protein_pr_water)
print("mlgr_cal_iron = ", mlgr_cal_iron)
```

```python
col_names = iffood.columns.tolist()
print(col_names)
```

```python
#creer un indice nutritionnel
# Score = 2 * (Protein_(g)) - 0.75 * (Lipid_Tot_(g))
protein = iffood["Protein_(g)"] * 2
fat = iffood["Lipid_Tot_(g)"] * 0.75
rating = protein - fat
print(rating)
```

```python
#normaliser des colonnes : valeur/max() de colonne
normalize_protein = iffood["Protein_(g)"]/iffood["Protein_(g)"]
normalize_fat = iffood["Lipid_Tot_(g)"]  / iffood["Lipid_Tot_(g)"].max()
print("normalize_protein = ", normalize_protein)
print("normalize_fat = ", normalize_fat)
```

```python
iffood["Normalize_protein"] = normalize_protein

iffood["Normalize_fat"] = normalize_fat
print(iffood.head(1))
```

```python
#créer nv colonne p/ indice nutriment normalise
iffood["Norm_intr_indice"] = 2 * (iffood["Normalize_protein"]) - 0.75 * (iffood["Normalize_fat"])
print(iffood.head(1))
```

```python
iffood.sort_values("Norm_intr_indice")
```
