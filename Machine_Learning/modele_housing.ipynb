```python
import pandas as pd

datapath = 'Jeux_de_donnees/'

# Load the data
housing = pd.read_csv(datapath + "housing.csv")
housing
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

modalite_ocean_proximity = pd.get_dummies(housing['ocean_proximity'])
housing = pd.concat([housing,modalite_ocean_proximity], axis=1) 
correlation = housing.corr()


```

```python
plt.figure(figsize=(20,15))
plt.subplot(111)
sns.heatmap(correlation, annot=True, linewidths=1.0)
plt.show()
```
