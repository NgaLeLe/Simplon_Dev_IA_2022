```python
#creer class s/balance
bank["y_num"] = bank[["y"]].apply(lambda x: 1 if x["y"] == "yes" else 0, axis=1)
bank["y_disc"] = pd.cut(bank["y_num"], bins=4, labels=["class1", "class2", "class3", "class4"])
bank["y_disc"].value_counts()
```

```python
bank["age_interval"] = bank[(bank["age"]<=25)].apply(lambda x: 25, axis=1) 
bank["age_interval"] = bank[(bank["age"]<=35) & (bank["age"]>25)].apply(lambda x: 35, axis=1) 
bank["age_interval"] = bank[(bank["age"]<=45) & (bank["age"]>35)].apply(lambda x: 45, axis=1) 
bank["age_interval"] = bank[(bank["age"]<=55) & (bank["age"]>45)].apply(lambda x: 55, axis=1) 
bank["age_interval"] = bank[(bank["age"]>55)].apply(lambda x: 56, axis=1) 
bank["age_interval"].value_counts()
```

```python
cont_age = bank[["age", "y_disc"]].pivot_table(
    index="age", columns="y_disc", aggfunc=len).fillna(0).copy().astype(int)
cont_age
```

```python
from scipy.stats import chi2_contingency
st_chi2, st_p, st_dof, st_exp = chi2_contingency(cont_age)
st_p
```
