```python
def creerListeRandom(n) :
    l = []
    for i in range(n) :
        l.append(randint(0,50))
    return l

def tableComparer(l1, l2) :
   #print liste 1 comme ligne 1
    print(" "*3, end="")
    for i in l1 :
        print("%4s" %i, end="")
    print("")
    #print chaque éle de liste 2 comme le nom de la ligne
    for j in l2 :
        print("%3s"%j, end="")
        for i in l1 :
            if j > i :
                signe = ">"
            elif j == i :
                signe = "="
            else :
                signe = "<"
            print("%4s" %(signe), end="")
        print("")

### MAIN ####
from random import randint
print("Ex8. Tableau de comparison 2 listes")
s1, s2 = 10, 6

li1 = creerListeRandom(s1)
li2 = creerListeRandom(s2)
print(li1)
print(li2)
print("\n---TABLE")
tableComparer(li1, li2)
```
