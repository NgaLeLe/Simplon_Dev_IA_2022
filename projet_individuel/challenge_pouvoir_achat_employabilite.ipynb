```python
def nb_etudiant_cat() :
    etudiant_cat = {}
    maj_category = recent_grads["Major_category"].unique()
    for cat in maj_category :
        etudiant_cat[cat] = recent_grads["Total"][recent_grads["Major_category"]==cat].sum()
    return etudiant_cat

nom_etu_cat = nb_etudiant_cat().
print("Nombre d'etudiant par category")
print(nom_etu_cat)
```

```python
#Nombre d'etudiants par category
def nombre_etudiant_cat_dic(dataset) :
    count_cat = {}
    maj_category = dataset["Major_category"].unique()
    for cat in maj_category :
        count_cat[cat] = dataset["Total"][dataset["Major_category"]==cat].sum()
    return count_cat

aa_car_counts = nombre_etudiant_cat_dic(all_ages)
rg_car_counts = nombre_etudiant_cat_dic(recent_grads)
print("Nombre d'etudiants par category")
print("\n----- ALL AGES -----")
print(aa_car_counts)
print("\n----- RECENT GRADS -----")
print(rg_car_counts)
```

```python
import numpy as np
def pivot_nombre_etudiant_cat(dataset) :
    pivot = dataset.pivot_table(index = "Major_category", values = "Total", aggfunc=np.sum)
    return pivot

pivot_aa_car_counts = pivot_nombre_etudiant_cat(all_ages)
pivot_rg_car_counts = pivot_nombre_etudiant_cat(recent_grads)
print("\n----- PIVOT NOMBRE ETUDIANT/CATEGORY (ALL AGES)  -----")
print(pivot_aa_car_counts)
print("\n----- PIVOT NOMBRE ETUDIANT/CATEGORY (RECENT GRADS) -----")
print(pivot_rg_car_counts)
```

```python
#Taux de jobs à faible salaire
low_wage_proportion = recent_grads["Low_wage_jobs"].sum()/recent_grads["Total"].sum()
print("low_wage_proportion = {:.3%}".format(low_wage_proportion))
```

```python
#Taux de jobs à faible salaire PAR CATEGORY
def proportion_sal_faible_cat() :
    set_cat = {}
    ls_cat = recent_grads["Major_category"].unique()
    for cat in ls_cat :
        set_cat[cat] = round(100*(recent_grads["Low_wage_jobs"][recent_grads["Major_category"] == cat].sum())/(recent_grads["Total"][recent_grads["Major_category"]==cat].sum()),3)
    return set_cat

proportion_sal_cat = pd.DataFrame.from_dict(proportion_sal_faible_cat(), orient="index", columns=["% d'etudiant"] )
print("Propotion jeune diplome par category faible salary\n----------------------------------")
print(proportion_sal_cat)
```

```python
cat_by_all_ages = all_ages["Major"].unique()
cat_by_recent_grads = recent_grads["Major"].unique()
rg_lower_count = 0
for c in cat_by_recent_grads :
    unemp_rec = recent_grads["Unemployment_rate"][(recent_grads["Major"]==c) & (recent_grads["Unemployment_rate"].isnull() == False)].tolist()
    unemp_all = all_ages["Unemployment_rate"][(all_ages["Major"]==c) & (all_ages["Unemployment_rate"].isnull() == False)].tolist()
    if (len(unemp_rec) == 0 or len(unemp_all) == 0):
        continue
    if (len(unemp_rec) > 0) :
        nb_unemp_rec = unemp_rec[0]
    if (len(unemp_all) > 0) :
        nb_unemp_all = unemp_all[0]
    
    if (nb_unemp_rec < nb_unemp_all) : 
        rg_lower_count = rg_lower_count + 1

print("rg_lower_count = ", rg_lower_count)
```
