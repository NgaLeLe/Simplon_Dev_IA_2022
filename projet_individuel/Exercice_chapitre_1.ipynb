```python
#  Import de la librairi pandas avec l'allias pd

import pandas as pd
```

```python
#Création d'un dataframe

df = pd.DataFrame({'Colonne1': [1.1, 2.7, 5.3], 'Colonne2': [2, 10, 9], 'Colonne3': ['oui', 'non', 'non']},
                      index = ['a', 'b', 'c'])

print(df)
```

```python
# Import de données depuis un fichié Excel dans un dataframe

nom_colonnes = ['consommation', 'cylindres', 'deplacement',
                'puissance', 'poids', 'acceleration', 'annee', 'origine', 'nom']

data = pd.read_excel('Desktop\Auto-mpg.xlsx',
                     index_col=None, header=None, names=nom_colonnes)
```

```python
# Afficher les premières lignes du dataframe
print(data.head())
```

```python
# Afficher les dimesions du dataframe
print(data.shape)
print(data.shape[0])
print(data.shape[1])
```

```python
#informations sur les données
print(data.info())
```

```python
#description des données
print(data.describe())
```

```python
# Fréquences corisés
print(pd.crosstab(data['annee'],data['cylindres']))
```

```python
# Pourcantage corisés
print(pd.crosstab(data['annee'],data['cylindres'], normalize='index'))
```

```python
#accès à une ou plusieurs colonne
print(data[['nom', 'consommation']].head())
```

```python
# Affiche les valeurs 10 à 12
print(data['nom'][10:13])
```

```python
# Accées à la valeur de la première ligne, première colonne
print(data.iloc[0, 0])
# Accées à la valeur de la dernière ligne, première colonne
print(data.iloc[-1, 0])
```

```python
# Toutes les colonnes avec les ligne de 0 à 4
print(data.iloc[0:5,:])
```

```python
# Liste des voitures dont son de l'année 80
print(data.loc[data['annee']==80,:].head())
```
