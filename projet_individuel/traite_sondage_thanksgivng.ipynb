```python
#creer un df contenue main_dish et type_cooked
df_maindish_typecook = data[['What is typically the main dish at your Thanksgiving dinner?','How is the main dish typically cooked?']]

#supprime ligne dont la valuer du colonne 'How is the main dish typically cooked type' est null(vide)
#df_maindish_typecook = df_maindish_typecook[df_maindish_typecook['How is the main dish typically cooked?'].isnull() == False]
df_maindish_typecook = df_maindish_typecook.dropna(axis = 0, subset = ['How is the main dish typically cooked?'])

#recupere ls valeurs de colonne 'How is...cooked type?'
typecook = df_maindish_typecook['How is the main dish typically cooked?'].unique().tolist()

#cree un nv colonne dont la valeur par défault est 0
df_maindish_typecook['type_cooked']=0
#modifie la valeur du colonne 'type_cooked' en focntion de la valeur du colonne 'How is...cooked type?'
i=0
for t in typecook :
    i += 1
    #localise les lignes par condition : loc[condition,colonne_vise]
    df_maindish_typecook.loc[df_maindish_typecook['How is the main dish typically cooked?'] == t, 'type_cooked'] = i

#cree un pivot qui affiche nb main_dish par type_cooked
pivot_maindish_typecooked = pd.pivot_table(df_maindish_typecook, index = 'What is typically the main dish at your Thanksgiving dinner?',
                                          values = 'type_cooked', columns= 'How is the main dish typically cooked?', aggfunc=len )
pivot_maindish_typecooked
```
